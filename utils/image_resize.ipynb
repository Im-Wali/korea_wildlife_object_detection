```python
from PIL import Image
import glob
import os
import shutil

# 입력 상위 폴더 설정
input_folder_images = "C:/DEV/91. Dataset/175.야생동물 활동 영상 데이터(YOLO형태)/validation/images"
input_folder_labels = "C:/DEV/91. Dataset/175.야생동물 활동 영상 데이터(YOLO형태)/validation/labels"
input_folder_json = "C:/DEV/91. Dataset/175.야생동물 활동 영상 데이터(YOLO형태)/validation/json"

# 출력 상위 폴더 설정
output_folder_images = "C:/DEV/91. Dataset/175.야생동물 활동 영상 데이터(YOLO형태)_resize/validation/images"
output_folder_labels = "C:/DEV/91. Dataset/175.야생동물 활동 영상 데이터(YOLO형태)_resize/validation/labels"
output_folder_json = "C:/DEV/91. Dataset/175.야생동물 활동 영상 데이터(YOLO형태)_resize/validation/json"


# 이미지 파일 리사이즈 및 저장
for img_path in glob.glob(f"{input_folder_images}/**/*.jpg", recursive=True):
    try:
        relative_path = os.path.relpath(img_path, input_folder_images)
        output_path = os.path.join(output_folder_images, relative_path)

        # 이미지 파일이 이미 존재하면 다음으로 넘어감
#         print(output_path)
        if os.path.exists(output_path):
            continue

        with Image.open(img_path) as img:
            resized_img = img.resize((640, 640))  # 원하는 해상도
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            resized_img.save(output_path)
    except Exception as e:
        print(f"이미지를 처리할 수 없습니다: {img_path} - {e}")
        continue

# 라벨 파일 복사
for label_path in glob.glob(f"{input_folder_labels}/**/*.txt", recursive=True):
    relative_path = os.path.relpath(label_path, input_folder_labels)
    output_path = os.path.join(output_folder_labels, relative_path)

    # 라벨 파일이 이미 존재하면 다음으로 넘어감
    if os.path.exists(output_path):
        continue

    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    shutil.copy(label_path, output_path)

# JSON 파일 복사
for json_path in glob.glob(f"{input_folder_json}/**/*.json", recursive=True):
    relative_path = os.path.relpath(json_path, input_folder_json)
    output_path = os.path.join(output_folder_json, relative_path)

    # JSON 파일이 이미 존재하면 다음으로 넘어감
    if os.path.exists(output_path):
        continue

    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    shutil.copy(json_path, output_path)

```
